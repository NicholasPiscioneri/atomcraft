## Toroidal Field Coils

All of the inner TF legs pass through a cylinder with radius $r_1$. Each coil is then modelled as a Princeton Dee curve with inner and outer radiii $r_1$ and $r_2$ respectively, such that $r_1 r_2 = r_0^2$. The shape factor is
$$
k = \frac{1}{2}\ln\left(\frac{r_2}{r_1}\right)
$$

There are $N_{TF}$ coils, each with $N_w$ windings, which each carry a current $i_{TF}$. For clarity, we denote the total current in capitals $I_{TF}= N_{TF} N_w i_{TF}$.

The coil voltage is given by
$$
\frac{v_{TF}}{N_{TF} N_w} = V_{TF} = R_{TF} I_{TF} + L_{TF} \frac{dI_{TF}}{dt}
$$
where
$$
R_{TF} = \frac{2 k r_0 (I_0(k) + I_1(k)) \eta}{f\epsilon r_1^2}
$$
and
$$
L_{TF} = 2\mu_0\pi k^2 r_0 (I_0(k) + 2 I_1(k) + I_2(k))
$$
$\eta$ is the resitivity of copper, $f$ is the circle packing fill factor, $\epsilon$ is the percentage volume of copper per wire and $I_n(x)$ is the modified Bessel function of the first kind.

From Ampere's Law, the maximum current in the coils over the pulse is
$$
I_{TF,max} = \frac{2\pi r_0 B_0}{\mu_0}
$$

## Central Solenoid

The central solenoid is modelled as a cylinder with radius $r_{CS}$. height $h$ and thickness $d$. There are $N_{CS}$ total windings, each carrying a current $i_{CS}$. For clarity, we denote the total current in capitals $I_{CS}= N_{CS} i_{CS}$. We assume that the central solenoid current swings from $+I_{max}$ to $-I_{max}$ over the course of the pulse.

The coil voltage is given by
$$
\frac{v_{CS}}{N_{CS}} = V_{CS} = R_{CS} I_{CS} + L_{CS} \frac{dI_{CS}}{dt}
$$
where
$$
R_{CS} = \frac{2\pi r_{CS} \eta}{f\epsilon h d}
$$
and
$$
L_{CS} = \frac{\mu_0 \pi r_{CS}^2}{h}
$$

### Estimating the Breakdown Discharge Rate

For breakdown, we are aiming for an electric field of ~1 V/m over 10 ms. These values will change as we build our model in DYON. From Faraday's law, the discharge rate during breakdown is given by
$$
\frac{dI_{CS,br}}{dt} = \frac{2 r_0  h E_{br}}{\mu_0 r_{CS}^2}
$$

### Estimating the Flat-Top Discharge Rate 

The circuit equation for the plasma is
$$
V_p = 0 = R_p I_p + L_p \frac{dI_p}{dt} - M \frac{dI_{CS}}{dt}
$$
If we assume that the plasma current is steady state during the flat-top stage, then
$$
0 = R_p I_p - M \frac{dI_{CS}}{dt}
$$
And therefore
$$
\frac{dI_{CS,ft}}{dt} = \frac{R_p I_p}{M}
$$
The mutual inductance between the solenoid and plasma is approximated as two coplanar concetric rings of current
$$
M = \frac{\mu_0\pi r_{CS}^2}{2 r_0}
$$
The plasma resistance is approximated using the Spitzer resitivity
$$
R_p = \frac{2\pi r_0 \eta_{Sp}}{\kappa\pi a^2} = 1.03\times10^{-4} \frac{Z \ln\Lambda}{T_e^{3/2}}\frac{r_0}{\kappa a^2}
$$
where $Z$ is the ionization, $\ln\Lambda$ is the Coulomb logarithm, $T_e$ is the electron temperature, $\kappa$ is the elongation and $a$ is the plasma minor radius. For our helium plasma, we take $Z\approx2$ and $\ln\Lambda\approx20$.

A more exact description of the plasma resistance will be obtained from both TokaMaker and DYON.

# Heating
For any coil, the energy dissipated is
$$
W = R \int_0^t I^2 dt
$$
The temperature rise is then
$$
\Delta T = ((c_{p,cu}\epsilon + c_{p,in}(1 - \epsilon)) fV)^{-1} W
$$
where $c_{p,cu}$ and $c_{p,in}$ are the isobaric volumetric heat capacities of the copper and insulation, respectively.
